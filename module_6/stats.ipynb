```python
import matplotlib.dates as mdates
#Task: Understand Museum attendance
museum_df = pd.read_csv('../data/museum_visitors.csv', index_col='Date', parse_dates=True)
museum_df
plt.figure(figsize=(16,6))
ax = sns.lineplot(data=museum_df)
plt.title("Museum Attendance from 2014-2019")
plt.ylabel('Attendance')
# Set x-axis to show every 2 days
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m%y'))
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Replace inf with NaN
import numpy as np
museum_df.isnull().sum()
museum_df = museum_df.replace( [np.inf, -np.inf], np.nan)
# Now we can use the fillna, replace Nan with 0
museum_df = museum_df.fillna(0)
```

```python
# Assess Seasonality
# Part A - Create a line chart that shows how the numer of visitors to the Avila adobe has evolved over time
plt.figure(figsize=(16,6))
ax = sns.lineplot(data=museum_df['Avila Adobe'], label = 'Avila Adobe' )
plt.title("Museum Attendance from 2014-2019")
plt.ylabel('Attendance')
# Set x-axis to show every 2 days
ax.xaxis.set_major_locator(mdates.DayLocator(interval=60))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m%y'))
plt.xticks(rotation=45)
plt.tight_layout()
```

## Scatter Plots

### Load Dataset
We will use a synthetic dataset of medical insurance charges:

```python
med_df = pd.read_csv('../data/insurance.csv')
med_df
```

```python
# Seaborn scatterplot
sns.scatterplot(x = med_df['bmi'], y=med_df['charges'])
plt.title('BMI vs Charges')
```

Looks like we have a `positive correlation` between BMI and Charges

```python
# To Double check strength, add a regression line to the data
sns.regplot(x=med_df['bmi'], y=med_df['charges'])
```

### Color-coded scatter plots
They are good when you need to display the correlation between more than 2 variables.

We want to understand how `smokers` affect the relationship between BMI and Charges.



```python
# Scatterplot of BMI vs charges and AND Smoker vs charges

sns.scatterplot(x=med_df['bmi'], y=med_df['charges'], hue=med_df['smoker'])
```

```python
# Add two regression lines
# When using lmplot for 3 or more, only use the column name
sns.lmplot(data=med_df, x = 'bmi', y= 'charges', hue= 'smoker', )
```

Finally, we can adapt the design of the scatter plot to feature a `categorical variable` (like smoker) on one of the main axis.
This plot is called **Categorical Scatter Plot**.

```python
sns.swarmplot(x = med_df['smoker'], y = med_df['charges'])
```
