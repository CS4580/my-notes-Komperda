## Task: Game Platform Preference
Analyze dataset from IGN

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
ign_df = pd.read_csv('../data/ign_scores.csv', index_col='Platform')
ign_df
ign_sorted = ign_df['Racing'].sort_values(ascending=False)
ign_sorted
```

```python
# Task 1: Which Platform is best? 
# Create a bar chart that shows 
plt.figure(figsize=(16,6))
sns.barplot(x=ign_sorted.index, y = ign_sorted.values)
plt.title('Average Racing SCore by platform')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
# Task: All possible combinations
# Create a heatmap of the average scores by genre in each cell

sns.heatmap(ign_df,cmap="coolwarm", annot=True )
```

```python
# Get the actual number for best and worst
# Stack method
highest = ign_df.stack().idxmax()
print(highest)
highest_combination_value = ign_df.stack().max()
print(highest_combination_value)
lowest = ign_df.stack().idxmin()
lowest_combination_value = ign_df.stack().min()

```
