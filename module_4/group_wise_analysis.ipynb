## Groupby Analysis in Pandas



```python
import pandas as pd
```

```python
df = pd.read_csv('../data/winemag-data-130k-v2.csv')
df.head()
```

```python
# Create a group of reviews of the points of each wine

df.groupby('points')['points'].count()

# value_counts() is a shortcut to the groupby()
```

```python
# Get the cheapest wine in each point category
df.groupby('points')['price'].agg({'min','max','count'})
```

```python
# Use the apply() method
# So what did this code do??
df.groupby('winery').apply(lambda df: df.title.iloc[0])
```

```python
# More finegrained control (filtering) you can group by more than one column
# Get the best wine by country AND province

df.groupby(['country','province']).apply(lambda df:  df.loc[df.points.idxmax()])

```

```python
# Use the agg() function to run a lot of different function in your DataFrame
agg = df.groupby(['country'])['price'].agg(['sum'])
print(agg)
```

## Multi-indexes
# multi-index differs from regular index in that it has multiple levels


```python
country_reviewed = df.groupby(['country','province'])['description'].agg([len])
mi = country_reviewed.index # Check how many index you have in df
print(type(mi))
```

```python
# You need to reset_index and add it back into multi-index df
country_reviewed.reset_index()
```

```python
# sort by length instead of default

country_reviewed
# sort is ascending by default, pass ascending = False parameter
country_reviewed.sort_values(by='len', ascending=False)
```
