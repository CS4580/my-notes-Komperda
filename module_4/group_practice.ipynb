```python
import pandas as pd
# Load Data
reviews_df = pd.read_csv('../data/winemag-data-130k-v2.csv')
reviews_df.head()
```

```python
reviews_df['taster_name'].value_counts()
```

```python
#Todo: Who are the most common wine reviewers
# Create a series whose index is the 'taster_twitter_handle'
# this does the same thing that .size() does
reviews_df.groupby('taster_twitter_handle')['taster_name'].count().sort_values(ascending=False)
```

```python
#TODO 2: What is the best wine I can buy for a given amount of money?
# Create a series whose index is the wine price and whose values are the max number of points
best_rating = reviews_df.groupby('price')[['points','title']].max().sort_index()
best_rating
```

```python
#TODO: What are the minimum and maximum prices for each variety of wine.
# index = variety, value are min and max
variety_price= reviews_df.groupby('variety')['price'].agg(['min','max'])
variety_price
```

```python
#TODO: What are the most expensive wine varieties?
# Create a variable 'sort_varieties' containing a copy of
# the dataframe from the previous task.
# so this is different than just sorting by max?
sort_varieties = variety_price.sort_values(by=['min', 'max'], ascending=False)
sort_varieties
#sorted_varieties = reviews_df.groupby('variety')['price'].max().sort_values(ascending=False)
```
