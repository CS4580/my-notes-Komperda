## Identify Missing Values

````python
# Find the number of missing values
df_raw.isnull().sum()
````

````python
# Missing values by percentage
(df_raw.isnull().sum() / len(df_raw)) * 100
````

````python
# Remove the company column/feature, keep the same df, don't copy
df_raw.drop('company', inplace=True)
````

````python
df_raw.info()
````

````python
# How to decidie what to fill missing values with
# Could do Mean, or Median, or Mode, or Zeros, or Forward Fill, or Backward Fill (use previous record value) (dangerous because of outliers)
# RandomFill -> within range
# Most common are Zero, Mean, Median
````

````python
# FIll in missing values in Children feature with zeros

df_raw.children = df_raw.children.fillna(0)
````

````python
print(df_raw['children'].isnull().sum())
df_raw.isnull().sum()
````

````python
# Unique values in the country feature
df_raw['country'].value_counts()
````

````python
# Drop all rows with missing data values in the country feature
df_raw = df_raw.dropna(subset=['country'])

````

````python
(df_raw.isnull().sum() / len(df_raw)) * 100
````
