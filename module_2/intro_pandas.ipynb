```python
import pandas as pd
```

```python
# Create a new dataFrame
df_one = pd.DataFrame(columns= ['Column1', 'Column2'])
df_one
```

```python
# Another way
df_one.columns = ['First', 'Second']
df_one
```

```python
# Add some rows
data1 = {'First':[1,2,'Real',],
         'Second':[22,44, 'Madrid']} # key = column, values = list of data

df_two = pd.DataFrame(data1) # Indexed like a regular array on the rows
df_two
```

```python
# Combined two dfs
df_three = pd.concat([df_one,df_two])
df_three
```

```python
# Add data to df_one
data = {
    'First': [66,77,'Hello'],
    'Second': [88,99,'CS4580']
}

# Add it to the frame
df_one = pd.DataFrame(data)
df_three = pd.concat([df_one,df_two])
df_three                  # Indexes are messed up.
```

```python
# Update indexes as you combine them
df_three.reset_index(drop=True, inplace=True)
df_three
```

```python
# Add df column-wise

col_data = {'Third':[88,99,11]}
df_four = pd.DataFrame(col_data)
# combine by column, add the axis = 1 parameter - what does this parameter mean exactly?
df_combined = pd.concat([df_three, df_four], axis=1)
df_combined
```
